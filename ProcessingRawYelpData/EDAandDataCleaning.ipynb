What this notebook does:

Exploratory data analysis of the candidate coffee shops obtained using the Yelp API. No review information is used. It is not yet verified that the candidate coffee shops are indeed coffee shops.  These were simply the businesses returned by the Yelp fusion api with the 'coffee' keyword.

Geographic distribution as well as distributions of the Yelp rating, number of reviews, number of locations of the businesses are tabulated and visualized.

The table of shops is annotated with data related to the neighborhood tabulation area of the shop. The Yelp category field is processed to make it easier to extract Yelp categories for each shop. The augmented table is saved in './ProcessedData/shops_withcategoriesgeoms.csv'

```python
m_1 = folium.Map(location=[40.74,-73.98],tiles='cartodbpositron',zoom_start=13)

#add shops to the map, color coded by price

def color_producer(val):
    if val == '$':
        return 'yellow'
    if val == '$$':
        return 'darkred'
    if val == '$$$':
        return 'black'
    if val == '$$$$':
        return 'blue'
    
for i in range(0,len(shops)):
    Circle(
        location=[shops.iloc[i]['latitude'],shops.iloc[i]['longitude']],
        radius = 20, 
        color = color_producer(shops.iloc[i]['price'])).add_to(m_1)
    
#display the map
m_1

```
